```python
import numpy as np
import statsmodels.api as sm
import statsmodels.graphics.tsaplots as tsaplots
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
import warnings

warnings.filterwarnings("ignore")
```

# Gaussian white noise

```python
def xxx(ar,ma,nsamp=1000,figsize=(15,5)):
    process = sm.tsa.ArmaProcess(ar=ar,ma=ma) 
    z = process.generate_sample(nsample=nsamp)
    plt.figure(figsize=figsize)
    plt.plot(z,linewidth=2,marker='o',color='darkslategray',
             markerfacecolor='cyan',
             markersize=4)
    plt.xlim([0,nsamp])
    plt.grid()

ar=[0.4,0.1]
ma=[0.9,0.4]
xxx(ar,ma,nsamp=200,figsize=(15,4))
```

# Sample autocorrelation

```python
ar=np.array([0.4,0.1])
ma=np.array([0.9,0.4])
process = sm.tsa.ArmaProcess(ar=ar,ma=ma)
plot_acf(process.acf(), lags=20);
```

```python
nsamp = 1000
z = process.generate_sample(nsample=nsamp)
plt.figure(figsize=(12,5))
plt.plot(z,linewidth=1)
plt.xlim([0,nsamp])
plt.grid()
```

```python
fig = plt.figure(figsize=(5,5))
plt.subplot()
tsaplots.plot_acf(z,lags=20, ax=fig.axes[0], title='')
plt.show()
```
