```python
import statsmodels.graphics.tsaplots as tsaplots
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import warnings

warnings.filterwarnings("ignore")
```

```python
nsamp = 500
figsize1 = (5,5)
figsize2 = (7,5)
```

```python
def doit(ma, ar, acforpcf):
    process = sm.tsa.ArmaProcess(ar,ma)
    z = process.generate_sample(nsample=nsamp)

    plt.figure(figsize=figsize2)
    plt.plot(z,linewidth=2,marker='.')
    plt.xlim([0,nsamp])
    plt.grid()
    plt.show()
    
    plot_acf(process.acf(), lags=20);

    fig = plt.figure(figsize=figsize1)
    plt.subplot()
    tsaplots.plot_acf(z,lags=20, ax=fig.axes[0], title='')
    plt.show()

    plot_pacf(process.pacf(), lags=20);

    fig = plt.figure(figsize=figsize1)
    plt.subplot()
    tsaplots.plot_pacf(z,lags=20, ax=fig.axes[0], title='')
    plt.show()
```

# MA(2)

```python
ma = [0.5, 0.2]
ar = [0.4,0.1]
z = doit(ma, ar, 'acf')
```

# MA(5)

```python
ma = [0.5, 0.5, 0.3, 0.2, 0.2]
ar = [0.4,0.1]

doit(ma, ar, 'acf')
```

```python
ma = [0.4, 0.4, 0.4, 0.4, 0.4, 0.4, 0.4, 0.4, 0.4, 0.4]
ar = [0.4,0.1]

doit(ma, ar, 'acf')
```
