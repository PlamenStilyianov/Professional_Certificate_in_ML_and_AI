```python
import graphviz
from sklearn import tree
from sklearn.datasets import load_iris

iris = load_iris()
X, y = iris.data, iris.target
dec_tree_model = tree.DecisionTreeClassifier(criterion='entropy')
model = dec_tree_model.fit(X,y)

data = tree.export_graphviz(model, out_file=None,
                                feature_names=["petal_length", "petal_width", "sepal_length", "sepal_width"],
                                class_names=["setosa", "versicolor", "virginica"],
                                filled=True, rounded=True,
                                special_characters=True)
graph = graphviz.Source(data)
graph
```

```python
import graphviz
from sklearn import tree
from sklearn.datasets import load_iris

iris = load_iris()
X, y = iris.data, iris.target
dec_tree_model = tree.DecisionTreeClassifier(criterion='gini')
model = dec_tree_model.fit(X,y)

data = tree.export_graphviz(model, out_file=None,
                            feature_names=["petal_length", "petal_width", "sepal_length", "sepal_width"],
                            class_names=["setosa", "versicolor", "virginica"],
                            filled=True, rounded=True,
                            special_characters=True)
graph = graphviz.Source(data)
graph
```
